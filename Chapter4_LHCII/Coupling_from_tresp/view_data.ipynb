```python
#from natsort import natsorted, ns
import pickle
#import csv
import numpy as np
#from load_mol2_data import *
import os
from math import *
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def norm_data(data):
    result = []
    for i in data:
        result.append((float(i)-min(data))/(max(data)-min(data)))
    return result
```

```python
def mirror_data(data):
    result = []
    for points in data:
        points = points + points.T - np.diag(np.diag(points))
        result.append(points)
    #print(np.asarray(result).shape)
    return np.asarray(result)
```

```python
def diag_data(data):
    results = {"eigenvalues":[], "eigenvectors": []}
    for points in data:
        values, vectors = np.linalg.eig(points)
        results["eigenvalues"].append(values)
        results["eigenvectors"].append(vectors)
    return results
```

```python
def square_mat(data):
    result = []
    for points in data:
        points = np.square(points)
        result.append(points)
    return np.asarray(result)
```

```python
def plot_contribution(vector,name1,name2):
    if name1 == "LUT620":
        a = 0
    elif name1 == "CLA610":
        a = 1
    elif name1 == "CLA611":
        a = 2
    elif name1 == "CLA612":
        a = 3
    if name2 == "LUT620":
        b = 0
    elif name2 == "CLA610":
        b = 1
    elif name2 == "CLA611":
        b = 2
    elif name2 == "CLA612":
        b = 3
    plt.plot(np.arange(0,1000), abs(vector[:,a,b]), 'ro')
    plt.axis([0, 1000, 0, 1])
    plt.show()
    return
```
